```python
import pandas as pd
from pathlib import Path
import seaborn as sns
from utils import get_ngram_counts
```

```python
reviews = pd.read_csv(
    Path("data") / "reviews_with_lemmas.csv",
    dtype={
        "appId": str,
        "language": str,
        "reviewId": str,
        "score": "int8",
        "content": str,
        "thumbsUpCount": "int32",
        "reviewCreatedVersion": str,
        "replyContent": str,
        "appVersion": str,
        "lemmas": str,
    },
    parse_dates=["at", "repliedAt"],
)
apps_ids = reviews["appId"].unique()
languages = reviews["language"].unique()
```

```python
reviews.info()
```

```python
filtered_reviews = reviews[reviews["score"] <= 2]
filtered_reviews.info()
```

```python
filtered_reviews["lemmas"] = filtered_reviews["lemmas"].fillna("")

reviews_data_grouped = {}
for app_id in apps_ids:
    app_reviews = filtered_reviews[filtered_reviews["appId"] == app_id]
    app_ngrams = get_ngram_counts(app_reviews)
    reviews_data_grouped[app_id] = {"reviews": app_reviews, "ngrams": app_ngrams}
```
